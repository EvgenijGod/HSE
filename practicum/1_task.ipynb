```python
import numpy as np
from skimage.io import imread, imshow, imsave
from skimage import img_as_float
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = imread('img.png')
number_of_rows = img.shape[0]
img = img_as_float(img)
img.dtype
```

```python
plt.figure(figsize=(10,10))
imshow(img)
```

```python
height = number_of_rows // 3
```

```python
img_blue = img[:height,]
plt.imshow(img_blue,cmap='gray')
```

```python
img_green = img[height:height*2,]
plt.imshow(img_green,cmap='gray')
```

```python
img_red = img[height*2:height*3,]
plt.imshow(img_red,cmap='gray')
```

```python
vert_edge = int(height * 0.1)
hor_edge = int(img.shape[1] * 0.1)
print(vert_edge, hor_edge)
```

```python
img_blue = img_blue[vert_edge:-vert_edge,hor_edge:-hor_edge]
plt.imshow(img_blue,cmap='gray')
```

```python
img_green = img_green[vert_edge:-vert_edge,hor_edge:-hor_edge]
plt.imshow(img_green,cmap='gray')
```

```python
img_red = img_red[vert_edge:-vert_edge,hor_edge:-hor_edge]
plt.imshow(img_red,cmap='gray')
```

```python
dest = np.dstack((img_red, img_green, img_blue))
imshow(dest)
```

```python
img_blue_c = np.roll(img_blue, 15, axis=0)
#imshow(img_blue,cmap='gray')
imshow(img_blue_c,cmap='gray')
```
